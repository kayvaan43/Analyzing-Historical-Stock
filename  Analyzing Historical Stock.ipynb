### Question 5

```python
import yfinance as yf
import requests
from bs4 import BeautifulSoup
import pandas as pd
import plotly.graph_objects as go

# Descargar datos de acciones de Tesla
tesla_ticker = 'TSLA'
tesla_stock_data = yf.download(tesla_ticker, start='2020-01-01', end='2023-01-01')

# Extraer datos de ingresos de Tesla
tesla_revenue_url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Referer': 'https://www.google.com/',
}
response = requests.get(tesla_revenue_url, headers=headers)
response.raise_for_status()
soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find('table', {'class': 'historical_data_table table'})

tesla_revenue_data = []
if table:
    for row in table.find_all('tr')[1:]:
        cells = row.find_all('td')
        if len(cells) > 0:
            year = cells[0].text.strip()
            revenue = cells[1].text.strip().replace('$', '').replace(',', '')
            tesla_revenue_data.append([year, int(revenue)])

tesla_revenue_df = pd.DataFrame(tesla_revenue_data, columns=['Year', 'Revenue'])

# Crear gráfico de precios de acciones de Tesla
fig_stock = go.Figure()
fig_stock.add_trace(go.Scatter(x=tesla_stock_data.index, y=tesla_stock_data['Close'],
                               mode='lines', name='Tesla Stock Price'))
fig_stock.update_layout(title='Tesla Stock Price',
                        xaxis_title='Date',
                        yaxis_title='Stock Price (USD)')

# Crear gráfico de ingresos de Tesla
fig_revenue = go.Figure()
fig_revenue.add_trace(go.Bar(x=tesla_revenue_df['Year'], y=tesla_revenue_df['Revenue'],
                             name='Tesla Revenue'))
fig_revenue.update_layout(title='Tesla Revenue',
                          xaxis_title='Year',
                          yaxis_title='Revenue (Millions USD)')

# Mostrar los gráficos
fig_stock.show()
fig_revenue.show()
```
